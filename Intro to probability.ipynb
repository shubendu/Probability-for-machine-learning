```python
import pandas as pd
import numpy as np
df = pd.read_csv('loan.csv')
df.head()
```

## In this task we will find the probability that if a person applied for a loan what is the probability that his application is approved.

```python
# Display values per class of 'Loan_Status'
print(df['Loan_Status'].value_counts())

# Total number of rows
total = df['Loan_Status'].value_counts().sum()

# Number of loan applications accepted
yes = df['Loan_Status'].value_counts().Y

# Number of loan applications rejected
no = df['Loan_Status'].value_counts().N

# Probability of loan getting accepted
p_of_A = yes/total

# Display probability
print(p_of_A)
```

## If you take a closer look at the data Property_Area is classified into three categories viz. Urban, Semi-Urban and Rural. In this task let us try to find out the probability that a randomly selected house is from either an urban area or a semi-urban area.

```python
df.head(10)
```

```python
# Total number of records
total = len(df)

# Probability of Urban houses
p_of_urban = len(df[df['Property_Area'] == 'Urban']) / total

# Probability of Semiurban houses
p_of_semiurban = len(df[df['Property_Area'] == 'Semiurban']) / total

# Probability of either Urban or Semiurban houses 
p_of_or = p_of_urban + p_of_semiurban

# Display probability
print(p_of_or)
```

## In this task let us check whether the gender of an applicant affects the chances of getting a loan. We will calculate the conditional probability in this case. We will calculate the probability of laon being approved when we know that the applicant is male (and female).

```python
# Convert to NumPy arrays
g = np.array(df.Gender)
l = np.array(df.Loan_Status)

# creating pivot table 
table = pd.crosstab(g,l)
print(table)

# Total male applicants
male = table.iloc[1,:].sum()

# Total female applicants
female = table.iloc[0,:].sum()

# Total male applicants whose loan applications were accepted
yes_male = table.iloc[1,1]

# Total female applicants whose loan applications were accepted
yes_female = table.iloc[0,1]

# Probability of loan approval when applicant is male
p_of_ymale = yes_male/male

# Probability of loan approval when applicant is female
p_of_yfemale = yes_female/female
print(p_of_ymale,'and' ,p_of_yfemale)
```
